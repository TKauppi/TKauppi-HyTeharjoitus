```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```
dat = pd.read_csv("https://raw.githubusercontent.com//EsaLoh/HyTe_koulutus/master/Kapeamuoto_1_10.csv", encoding='latin-1')

dat[0:150] #tulostetaan otsikko ja ensimmäinen rivi tarkistuksen vuoksi
```

```
dat2 = dat[dat.Kysymys=='Nousin aamulla reippaasti sangysta ylos']
dat2
```

```
sns.set(style="whitegrid", rc={'figure.figsize':(16,8)})

ax = sns.barplot(x=dat.Nimimerkki, y=dat2.Arvo, data=dat2)
```

```
dat3 = dat[dat.Nimimerkki==7968]
dat3
```

```
sns.set(style="whitegrid", rc={'figure.figsize':(40,16)})

ax = sns.barplot(x=dat.Kysymys, y=dat3.Arvo, data=dat3)
```

```
fiilis = dat[dat.Kysymys=='Nukuin riittavasti viime yona']
teema = dat[dat.Kysymys=='Tunnen itseni onnelliseksi']

Fiilis = pd.Series(fiilis.Arvo)
Teema = pd.Series(teema.Arvo)

from numpy.polynomial.polynomial import polyfit
b, m = polyfit(Fiilis, Teema, 1)

plt.plot(Fiilis, Teema, '.')
plt.plot(Fiilis, b + m * Fiilis, '-')
plt.title("Teema vs Fiilis")
plt.show()
#plt.scatter(Fiilis, Teema)
```
